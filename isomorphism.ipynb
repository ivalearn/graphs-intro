## Доказать изоморфность графов

[Задача](https://stepik.org/lesson/10897/step/8?unit=2335)
решается подбором маркировки вершин левого и правого графа.

Пометим вершины левого буквами $A,B,C,D,E,F,G,H$, как показано на рисунке внизу слева.

Пометим вершины правого графа тем же набором букв, но снабдив их нижним индексом $1$.

Выпишем списки смежности для левого графа:
$A \rightarrow {B,E,D}$;
$B \rightarrow {A,F,C}$;
$F \rightarrow {B,E,G}$;
$E \rightarrow {A,F,H}$;
$C -> {B,D,G}$;
$D -> {A,C,H}$;
$H -> {E,D,G}$;
$G -> {F,C,H}.$

Запись списка смежности $x \rightarrow {a,b,c}$ означает, что вершина $x$ соединяется с каждой из вершин неупорядоченного множества ${a,b,c}$ ребром.

Заметим, что списки смежности для вершин правого графа, совпадают со списками выше с точностью до индекса $1$.

Таким образом, взаимно однозначное отображение $A \rightarrow A_1$, ..., $H \rightarrow H_1$ вершин левого графа в вершины правого сохраняет свойста смежности. А значит, наши графы изоморфны.

## Доказать неизоморфность графов

[Задача](https://stepik.org/lesson/10897/step/9?unit=2335)

Левый граф является двудольным: как показано на рисунке внизу слева, если раскрасить его вершины в красный и синий цвета, любая красная вершина смежна только с синими и наоборот. Изоморфное преобразование вершин φφ обязано сохранить свойство двудольности. Действительно, если две красные вершины не смежны между собой, то их отображения в изоморфном графе тоже не могут быть смежны, по определению изоморфности. Однако, правый граф не двудольный, как показано на рисунке справа внизу. Даже если мы попытаемся чередовать красный и синий цвета при раскраске вершин на окрыжности, рёбра-радиусы соединят вершины одного цвета (такие, как $p8-p4$) из-за чередования четности вершин. Таким образом, два наших графа не могут быть изоморфными.
